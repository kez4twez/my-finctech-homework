```python
## FOR MARKERS, All went well, only thing I couldn't figure out was how to output only the menu items that aren't in the sales_data file.
## But all in all it was a super fun one and I am glad I had a crack!

# Import Path and csv
from pathlib import Path
import csv

#Initialise necessary lists and dicts
menu = []
sales = []
no_match = {}

# Set the file paths and open the files
menu_path = Path("Resources/menu_data.csv")
sales_path = Path("Resources/sales_data.csv")

# Opening and reading the menu data and appending to menu[]
with open (menu_path, 'r') as file:
    menu_reader = csv.reader(file, delimiter=',')
    # Read the header and skip it
    menu_header = next(menu_reader)
    
    for row in menu_reader:
        menu.append(row)   
        
# Opening and reading sales data and appending to sales[]
with open (sales_path, 'r') as file:
    sales_reader = csv.reader(file, delimiter=',')
    # Read the header and skip it
    sales_header = next(sales_reader)
    
    for row in sales_reader:
        sales.append(row)
#-----------------------------------#   
# Initialise report_dict to eventually hold 
report_dict = {}

# Loop through sales
for row in sales:#[:50]:
    # Declare variables and assign their corresponding value from sales[]
    sales_quantity = int(row[3])
    sales_menu_item = row[4]
    if sales_menu_item not in report_dict.keys():
        # Initialise report_dict with the correct keys
        report_dict[sales_menu_item] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0}
        
    # Create a nested loop to compare menu_item to sales_menu_item and retrieve data for report_dict{}   
    for line in menu:
        # Declare variables and assign their corresponding value from menu[]
        menu_item = line[0]
        menu_price = float(line[3])
        menu_cost = float(line[4])
        profit = menu_price - menu_cost
        
        # Match sales_menu_item and menu_item and capture count, revenue, cogs and profit
        if sales_menu_item == menu_item:
            report_dict[sales_menu_item]["01-count"] += sales_quantity
            report_dict[sales_menu_item]["02-revenue"] += menu_price * sales_quantity
            report_dict[sales_menu_item]["03-cogs"] += menu_cost * sales_quantity
            report_dict[sales_menu_item]["04-profit"] += profit * sales_quantity
        # Store no match items in a dictionary to later print out
        else: 
            no_match[menu_item] = [f"Sales menu item does not equal {menu_item}! NO MATCH!"]

# Write report_dict to an output.txt file
output_path = Path("output.txt")

with open (output_path, "w") as file:
    for key, value in report_dict.items():
        file.write('%s:%s\n' % (key, value))
    
        
# **Testing prints**
print("Ramen Sales Figures")
print ("------------------------------------------------------------------------------------------------------------------------------------------") 
for items in report_dict.items():
    print (items)
    print ("------------------------------------------------------------------------------------------------------------------------------------------")   
      
```
