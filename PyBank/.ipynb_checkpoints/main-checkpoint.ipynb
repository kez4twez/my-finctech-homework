```python
# HINT!! Probably going to need to put all of the data into a dictionary before you calculate the end results
# Because at the end you are going to need to print the corresponding day, for example you will need to print:
# "Greatest Increase in Profits: Feb-2012 ($1926159)", and the easiest way to do this is to loop through key value pairs and printf both

# Haven't figured it out yet, but using a dictionary may also make it easier to do some of the calculations, like perhaps the avg_change

# Import the pathlib and csv library
from pathlib import Path
import csv
from numpy import mean
# Set the filepath 
csv_path = Path('Resources/budget_data.csv')

# Initialise 'data' list to later store budget_data.csv 
data = []
data_dict = {}
monthly_change = []

# Initialise variables to be used in later loop
total_months = 0
net_pl = 0
avg_change = 0
largest_increase = 0
largest_increase_key = 0
largest_decrease = 0
largest_decrease_key = 0

# Open the csv file as an object
with open (csv_path, 'r') as file:
    
    # Pass in the csv file to the csv.reader() function
    csv_reader = csv.reader(file, delimiter=',')
    
    # Read the header
    header = next(csv_reader)
    
    # Loop over the data to calculate total_months, net_pl, avg_change, largest_increase, largest_decrease
    for row in csv_reader:
        # Printing of each row in loop to make sure csv_reader is working correctly
        #print (row)
        # Calculate total months
        total_months += 1
        # Calculate net_pl
        net_pl += int(row[1])
        # Take all months and corresponding PL and put intoa list data[], as well as key value pairs in data_dict{} for further calculations
        data.append(row[1])
        key, value = row[0], row[1]
        data_dict[key] = int(value) #Make sure that if you are going to use this dict to compare ints that you turn every value into an int
            
# Calculate largest_increase and largest_decrease from data_dict so you can print month and value
for key, value in data_dict.items():
    if largest_decrease == 0:
        largest_decrease = value
    elif value < largest_decrease:
        largest_decrease = value
        largest_decrease_key = key
    if largest_increase == 0:
        largest_increase = value
    elif value > largest_increase:
        largest_increase = value
        largest_increase_key = key
            
# Calculate MoM change and store in list to later work out avg_change
for i, element in enumerate(data):
    current_pnl = element
    if i < len(data) - 1:
        next_pnl = data[i+1]
        change = int(next_pnl) - int(current_pnl)
        monthly_change.append(change)  
        
# Calulate month on month change and store in avg_change variable
avg_change = mean(monthly_change)

output_file_path = Path("output.txt")

with open (output_file_path, 'w') as file:
    file.write(f"Financial Analysis\n--------\nTotal months: {total_months}\nTotal profit/loss: ${net_pl}\n--------\nLargest increase in profits: {largest_increase_key} ${largest_increase}\nLargest decrease in profits: {largest_decrease_key} ${largest_decrease}\nAverage change: ${round(avg_change, 2)}")
    
# Printing of all figures formatted
print ("Financial Analysis")
print ("---------")
print (f"Total months: {total_months}")
print (f"Total profit/loss: ${net_pl}")

print ("---------")
print (f"Largest increase in profits: {largest_increase_key} ${largest_increase}")
print (f"Largest decrease in profits: {largest_decrease_key} ${largest_decrease}")
print (f"Average change: ${round(avg_change, 2)}")

```
